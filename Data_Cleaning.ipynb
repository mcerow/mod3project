```python
import pandas as pd
import numpy as np
```

```python
whiskey = pd.read_csv('whisky_df.csv')
```

```python
whiskey.head()
```

```python
whiskey.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
```

```python
whiskey['name'] = whiskey['name'].map(lambda x: x.replace("\n", ""))
whiskey['name'].head()
```

```python
whiskey['price'].dtypes
```

```python
whiskey['price'] = whiskey['price'].astype(str)
whiskey['price'] = whiskey['price'].map(lambda x: x.strip('[]'))
whiskey['price'] = whiskey['price'].map(lambda x: x.lstrip("'cost-"))
whiskey['price'] = whiskey['price'].map(lambda x: x.rstrip("', 'spirit-cost'"))
whiskey['price'].head()
```

```python
whiskey.columns
```

```python
for i in whiskey['distillery_loc']:
    whiskey['country'] = (whiskey['distillery_loc'].str.split('//').str[1])
    whiskey['country'].fillna(whiskey['distillery_loc'], inplace=True)
    whiskey['country'] = whiskey['country'].map(lambda x: str(x).replace("\n", ""))
    whiskey['country'] = whiskey['country'].str.split()
    whiskey['country'] = whiskey['country'].str[-1]
whiskey['country'].value_counts()
```
